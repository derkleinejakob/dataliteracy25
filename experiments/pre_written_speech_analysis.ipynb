```python
import pandas as pd

import preamble
from src.constants import PATH_ALL_SPEECHES
```

```python
df = pd.read_parquet(PATH_ALL_SPEECHES)
```

```python
df_2014_to_2018 = df[(df["year"] >= 2014) & (df["year"] < 2018)]

df_2014_to_2018.groupby("date").size().plot.bar(figsize=(20, 6), title="Number of speeches per day")
None
```

```python
top_3_dates = df_2014_to_2018.groupby("date").size().sort_values(ascending=False).head(3)
print("Top 3 dates with highest number of speeches")
print(top_3_dates)
```
