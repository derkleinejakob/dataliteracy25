```python
from gensim.models import LdaMulticore
import os 

root = "data/lda"
n_topics = 40
folder_n_topics = str(n_topics)+"_topics"
folder_n_passes = "10"

lda_model = LdaMulticore.load(os.path.join(root, folder_n_topics, folder_n_passes, "model.model"))
```

```python
show_probs = False
k_words = 5 

for topic_index, topic in lda_model.show_topics(formatted=False, num_topics=n_topics):
    if show_probs: 
        print(f"Topic {topic_index}:", [f"{topic_words} {'%.3f' % topic_probs}" for topic_words, topic_probs in topic[:k_words]])
    else: 
        print(f"Topic {topic_index}:", [topic_words for topic_words, _ in topic[:k_words]])
```
