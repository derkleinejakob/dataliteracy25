```python
N_SPEECHES = 10e6
AVG_SPEECH_LEN_CHARS = 500 * 10 # very bad estimate: 500 words with 10 chars each
N_CHARS_PER_TOKEN = 5
N_OUT_TOKENS = 600 # assume maybe 300 words each 2 tokens, also bad estimation
N_ADDITIONAL_IN_TOKENS = 200 # assume maybe a prompt of 200 tokens, also bad estimation

BUDGET = 300
```

```python
model_prices = {
    "GPT-5 nano": {"in": 0.05, "out": 0.4}, 
    "GPT-5 mini": {"in": 0.25, "out": 2},
    "GPT-5": {"in": 1.25, "out": 10},
    "Gemini-2.5 pro": {"in": 2.5, "out": 15}, # assumed input tokens > 200.000 but thats probab wrong, then would be 50% cheaper 
    "Gemini-2.5 flash GA": {"in": 3, "out": 2.5}, #same
    # "": {"in": 0, "out": 0},
}
use_batched_processing = True # 24h completion time, but will be 50% cheaper
```

```python
in_tokens_total = N_SPEECHES * (AVG_SPEECH_LEN_CHARS / N_CHARS_PER_TOKEN + N_ADDITIONAL_IN_TOKENS)
out_tokens_total = N_SPEECHES * N_OUT_TOKENS

print(f"Total cost for {N_SPEECHES} speeches:\n")
for model in model_prices.keys(): 
    in_cost = model_prices[model]["in"] * (in_tokens_total) / 10e6 # price is per million tokens
    out_cost = model_prices[model]["out"] * (out_tokens_total) / 10e6
    
    total = in_cost + out_cost
    if use_batched_processing: 
        total /= 2
    print(f"{model}:  {total}$")
```

```python
in_tokens_per_speech =  (AVG_SPEECH_LEN_CHARS / N_CHARS_PER_TOKEN + N_ADDITIONAL_IN_TOKENS)
import math 

print(f"Max speeches for {BUDGET}$:\n")
for model in model_prices.keys():
    price_per_speech = (model_prices[model]["in"] / 10e6) * in_tokens_per_speech +  (model_prices[model]["out"] / 10e6) * N_OUT_TOKENS
    if use_batched_processing: 
        price_per_speech /= 2
        
    n_speeches = math.floor((BUDGET / price_per_speech) / 1000)
    print(f"{model}:  {n_speeches} thousand speeches")
```
